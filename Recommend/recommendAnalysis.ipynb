```python
def getRecommendations(recommend_data, cap=5):
    """"""
    ### create blacklist (beatmaps' in the users' top 50 plays)
    black_list_ids = []
    for play in recommend_data['user_data']['plays']:
        black_list_ids.append(play['beatmap_id'])
        
    # print(black_list_ids)
    
    rec_dict = {}
    
    for user_data in recommend_data['recommend_data']:
        for play in user_data['plays']:
            
            ### assigning variables
            beatmap_id = play['beatmap_id']
            pp = play['pp']
            mods = play['enabled_mods']
            max_combo = play['maxcombo']
            perfect = play['perfect']
            miss_count = play['countmiss']
            
            # manually getting the accuracy
            count_dict = {}
            counts = "count100 count300 count50 countgeki countkatu countmiss".split(' ')
            for count in counts:
                count_dict[count] = play[count]
            accuracy = getAccuracy(mode=0, count_dict=count_dict)
            
            if beatmap_id not in rec_dict.keys():
                rec_dict[beatmap_id] = defaultdict(list)
            else:
                pass
            
            data = [('pps', pp), ('mods', mods), ('max_combos', max_combo), ('perfects', perfect), ('misses', miss_count)]
            for name, value in data:
                rec_dict[beatmap_id][name].append(value)
            
            
            if type(rec_dict[beatmap_id]['count']) == int:
                rec_dict[beatmap_id]['count'] += 1
            else:
                rec_dict[beatmap_id]['count'] = 1
            
            rec_dict[beatmap_id] = dict(rec_dict[beatmap_id])
            
                
                
    return rec_dict
            
```
