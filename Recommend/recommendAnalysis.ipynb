```python
def getRecommendations(recommend_data, blacklist=True, cap=5):
    """
    @param blacklist (boolean) optional - dont include beatmaps in the users top X plays (x = 50)
    """
    ### create blacklist (beatmaps' in the users' top 50 plays)
    blacklist_ids = []
    for play in recommend_data['user_data']['plays']:
        blacklist_ids.append(play['beatmap_id'])
        
    recommend_list = []    
        
    for user_data in recommend_data['recommend_data']:
        for play in user_data['plays']:
            
            ### assigning variables
            beatmap_id = play['beatmap_id']
            if blacklist:
                if beatmap_id in blacklist_ids:
                    continue
            
            pp         = float(play['pp'])
            mods       = int(play['enabled_mods'])
            combo      = int(play['maxcombo'])
            perfect    = bool(int(play['perfect']))
            miss_count = int(play['countmiss'])
            
            # manually getting the accuracy
            count_dict = {}
            counts = "count100 count300 count50 countgeki countkatu countmiss".split(' ')
            for count in counts:
                count_dict[count] = play[count]
            accuracy = float(getAccuracy(mode=0, count_dict=count_dict))
     
            
            beatmap_ids = [i['beatmap_id'] for i in recommend_list]
            
            if beatmap_id in beatmap_ids:
                d_index = beatmap_ids.index(beatmap_id)
                d = recommend_list[d_index]
                d['count'] += 1
                del recommend_list[d_index]
            else:
                d = defaultdict(list)
                d['beatmap_id'] = beatmap_id
                d['count'] = 1
            
            data = [('pps', pp), ('mods', mods), ('combos', combo), ('perfects', perfect), ('misses', miss_count)]
            for name, value in data:
                d[name].append(value)
            
            recommend_list.append(dict(d))
            
                
                
    return sorted(recommend_list, key=itemgetter('count'), reverse=True)
                            
```

```python
r = getRecommendations(recommend_data)
pprint(r)
#count_sum = sum([i['count'] for i in r])
#print(count_sum)
```
